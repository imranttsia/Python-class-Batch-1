Choosing outfit depending on the weather and event type

> Add blockquote



```python
Temp=float(input("Enter the temperature in degree centigrade: "))
is_formal_event=input("Is it a formal event? (yes/no): ")


if Temp>25: #warm weather
    if is_formal_event=="yes":
      print("Wear Bright cloths (Suiting wear)")
    else:
      print("Wear Darks cloths(T-shit and pant)")

elif Temp>10 and Temp<=25: #cold weather
    if is_formal_event=="yes":
      print("Wear Dark cloths (Suiting wear)")
    else:
      print("Wear Bright cloths(T-shit and pant)")
else: #extreme cold weather
    print("We will not go outdoor")

```

Using a Discount Coupon - Checking coupon eligibility based on the item category and the purchase total

```python
cart_total=float(input("Enter the total amount of the cart: "))
item_purchase= input("Enter the item category purchased (Do purchase electronic (yes/no)): ")  #There is a discount on electronic items
membership_status=input("Are you a member of our loyalty program? (yes/no): ")

if cart_total>=20000:
  if item_purchase=="yes":
    print("Eligible for 20 percent discount")

  elif membership_status=="yes":
      print("Eligible for 10 percent discount")
else:
  print("No discount")
```
