Rewrite your pay computation to give the employee 1.5 times the hourly rate for hours worked above 40 hours

```python
# Input from the user
hours = float(input("Enter total hours worked: "))
hourly_rate = float(input("Enter hourly rate: "))

if hours <= 40:
    # Regular pay for 40 or fewer hours
    total_pay = hours * hourly_rate
else:
    # Regular pay for first 40 hours + Overtime pay for extra hours
    regular_pay = 40 * hourly_rate
    overtime_hours = hours - 40
    overtime_pay = overtime_hours * (1.5 * hourly_rate)
    total_pay = regular_pay + overtime_pay

# Calculate and display the pay

print(f"Total pay: ${total_pay:.2f}")
```

Write a program to prompt for a score between 0.0 and 1.0. If the score is out of range , print an error message. If the score between 0.0 and 1.0, print a grade using the following table: 
(Score >=0.9, Grade=A); (Score >=0.8, Grade=B); (Score >=0.7, Grade=C); (Score >=0.6, Grade=D); (Score <0.6, Grade=F) 

```python
# Prompt user for a score and compute grade
try:
    # Input from user
    score = float(input("Enter a score between 0.0 and 1.0: "))
    
    # Check if score is within the valid range
    if score < 0.0 or score > 1.0:
        print("Error: The score must be between 0.0 and 1.0.")
    else:
        # Determine grade based on score
        if score >= 0.9:
            grade = "A"
        elif score >= 0.8:
            grade = "B"
        elif score >= 0.7:
            grade = "C"
        elif score >= 0.6:
            grade = "D"
        else:
            grade = "F"
        
        print(f"Grade: {grade}")

except ValueError:
    # Handle non-numeric input
    print("Error: Please enter a valid numeric score.")
```

Ask the user to enter a four digit number and check whether  the sum of the first and the last digit is same as the sum of the second and the third digit

```python
# Prompt the user for a four-digit number
number = input("Enter a four-digit number: ")

# Check if the input is valid (must be exactly 4 digits)
if number.isdigit() and len(number) == 4:
    # Extract digits
    first_digit = int(number[0])
    second_digit = int(number[1])
    third_digit = int(number[2])
    last_digit = int(number[3])
    
    # Calculate sums
    sum_first_last = first_digit + last_digit
    sum_second_third = second_digit + third_digit
    
    # Check if the sums are equal
    if sum_first_last == sum_second_third:
        print("The sum of the first and last digit is equal to the sum of the second and third digit.")
    else:
        print("The sums are not equal.")
else:
    print("Error: Please enter a valid four-digit number.")
```

```python
# Prompt the user for a four-digit number
number = input("Enter a four-digit number: ")

# Check if the input is valid (must be exactly 4 digits)
if number.isdigit() and len(number) == 4:
    # Extract digits
    first_digit = int(number[0])
    second_digit = int(number[1])
    third_digit = int(number[2])
    last_digit = int(number[3])
    
    # Calculate sums
    sum_first_last = first_digit + last_digit
    sum_second_third = second_digit + third_digit
    
    # Check if the sums are equal
    if sum_first_last == sum_second_third:
        print("The sum of the first and last digit is equal to the sum of the second and third digit.")
        
        # Increment the second and third digits by 1
        new_second_digit = second_digit + 1
        new_third_digit = third_digit + 1
        
        # Form the new number
        new_number = f"{first_digit}{new_second_digit}{new_third_digit}{last_digit}"
        print(f"The new number with the second and third digits incremented by 1 is: {new_number}")
    else:
        print("The sums are not equal.")
else:
    print("Error: Please enter a valid four-digit number.")
```

```python
import math

# Prompt the user for the concentration of hydrogen ions
try:
    concentration = float(input("Enter the concentration of hydrogen ions in the solution (C): "))
    
    # Check if the concentration is valid (positive number)
    if concentration <= 0:
        print("Error: Concentration must be a positive number.")
    else:
        # Calculate the pH using the formula
        pH = -math.log10(concentration)
        
        # Determine acidity, basicity, or neutrality
        if pH > 7:
            print(f"The pH of the solution is {pH:.2f}. The solution is basic.")
        elif pH < 7:
            print(f"The pH of the solution is {pH:.2f}. The solution is acidic.")
        else:
            print(f"The pH of the solution is {pH:.2f}. The solution is neutral.")
except ValueError:
    print("Error: Please enter a valid numeric value for the concentration.")
```

Find whether a number enetered by user is divisible by 3 and 13

```python
# Prompt the user to enter a number
try:
    number = int(input("Enter a number: "))
    
    # Check divisibility by 3 and 13
    if number % 3 == 0 and number % 13 == 0:
        print(f"The number {number} is divisible by both 3 and 13.")
    else:
        print(f"The number {number} is NOT divisible by both 3 and 13.")
except ValueError:
    print("Error: Please enter a valid integer.")
```

Find whether a number entered by the user is a perfect square

```python
import math

# Prompt the user to enter a number
try:
    number = int(input("Enter a number: "))
    
    if number < 0:
        print(f"The number {number} is NOT a perfect square (negative numbers cannot be perfect squares).")
    else:
        # Find the square root and check if it's an integer
        square_root = math.sqrt(number)
        if square_root == int(square_root):
            print(f"The number {number} is a perfect square.")
        else:
            print(f"The number {number} is NOT a perfect square.")
except ValueError:
    print("Error: Please enter a valid integer.")
```

Ask a user to enter a string and find the alphanumeric charaters from the string

```python
# Prompt the user to enter a string
user_string = input("Enter a string: ")

# Initialize an empty string to store alphanumeric characters
alphanumeric_chars = ''

# Iterate through each character in the string
for char in user_string:
    if char.isalnum():  # Check if the character is alphanumeric
        alphanumeric_chars += char  # Add the character to the result
    else:
        pass  # Non-alphanumeric characters are ignored

# Display the result
if alphanumeric_chars:
    print(f"Alphanumeric characters from the string: {alphanumeric_chars}")
else:
    print("No alphanumeric characters found in the string.")
```

```python
# Prompt the user to enter a string
user_string = input("Enter a string: ")

# Initialize empty strings for alphanumeric characters and digits
alphanumeric_chars = ''
digits = ''

# Iterate through each character in the string
for char in user_string:
    if char.isalnum():  # Check if character is alphanumeric
        alphanumeric_chars += char
        if char.isdigit():  # Check if the alphanumeric character is a digit
            digits += char

# Display the results
if alphanumeric_chars:
    print(f"Alphanumeric characters from the string: {alphanumeric_chars}")
else:
    print("No alphanumeric characters found in the string.")

if digits:
    print(f"Digits in the string: {digits}")
else:
    print("No digits found in the string.")
```
