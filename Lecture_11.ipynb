**stop**

```python
def countdown(n):
  if n==0:
    print("Stop")
countdown(3)
```

```python
def countdown(n):
  if n==0:
    print("Stop")
  else:
    print(n)
(countdown(3))
```

```python
def countdown(n):
  if n==0:
    print("Stop")
    return
    print(n)
(countdown(3))
```

```python
def countdown(n):
  if n==0:
    print("Stop")
    return
  print(n)
(countdown(3))
```

```python
def countdown(n):
  if n==0:
    print("Stop")
    return
  print(n)
  countdown(n-1)

(countdown(3))
```
