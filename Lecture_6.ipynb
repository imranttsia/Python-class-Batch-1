Determine security permission

```python
person=input("Enter the designation name. Write s for student, t for teacher and p for principal:  ")
if person=="s":
    print("You have read permission")
if person=="t":
    print("You have read and write permission")
if person=="p":
    print("You have read, write and delete permission")
```

Input number and check whether it is divisible by 2, 3 and 5

```python
num=int(input("Enter the number: "))
if num%2==0:
    print("Number is divisible by 2")
if num%3==0:
    print("Number is divisible by 3")
if num%5==0:
    print("Number is divisible by 5")
```

Chacking a bank balance for a withdrawal

```python
current_balance=100000
withdrawal_amount=int(input("Enter the withdrawal amount: "))
if withdrawal_amount<=current_balance:
    print("Withdrawal successful")
    current_balance=current_balance-withdrawal_amount
    print("Current balance is: ",current_balance)
else:
    print("Withdrawal failed because of insufficient balance")
```

Grading System using elif

```python
per=int(input("Enter the percentage: "))
if per>=90:
    print("Grade A")
elif per>=80:
    print("Grade B")
elif per>=70:
    print("Grade C")
else:
    print("Grade D")
```

Time of Day greeting

```python
timings=int(input("Enter the time in 24 hours format: "))
if timings>=5 and timings<12:
    print("Good Morning")
elif timings>=12 and timings<17:
    print("Good Afternoon")
elif timings>=17 and timings<19:
    print("Good Evening")
else:
    print("Good Night")
```

Bank Account fee Calculation

```python
balance=int(input("Enter your account balance"))
fee=0
if balance < 5000:
    fee=1000
elif balance >= 5000 and balance < 10000:
    fee=500
else:
    fee=0

print("Transaction fee is: ",fee)
```

Comparing two numbers

```python
num1=int(input("Enter the first number: "))
num2=int(input("Enter the second number: "))
if num1>num2:
    print("First number is greater")
elif num1<num2:
    print("Second number is greater")
else:
    print("Both numbers are equal")
```

Discount Eligibility

```python
purchase_amount=int(input("Enter the purchase amount: "))
if purchase_amount>=10000:
    print("You are eligible for 20% discount")
    discount=purchase_amount*0.2
    print("Discount amount is: ",discount)

    print("Purchase discount price: ",purchase_amount*0.8)
elif purchase_amount>=7000 and purchase_amount<10000:
    print("You are eligible for 10% discount")
    discount=purchase_amount*0.1
    print("Discount amount is: ",discount)
    print("Purchase discount price: ",purchase_amount*0.9)
else:
    print("You are not eligible for discount")
```
