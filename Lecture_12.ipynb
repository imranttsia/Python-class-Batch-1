```python
num = int(input("Enter a non-negative number: "))

factorial = 1

# check if the number is negative, positive or zero
if num < 0:
   print("Sorry, factorial does not exist for negative numbers")
elif num == 0:
   print("The factorial of 0 is 1")
else:
  while num>0:
   factorial=factorial*num
   num-=1
print("The factorial of",num,"is",factorial)
```

## Implement Alarm Logic

### Subtask:
Complete the `while` loop in cell `dsWiDVxpbwP1` to simulate a morning alarm. This will involve prompting the user to 'wake up' or 'snooze', incrementing a snooze counter, and setting the `awake` variable to `True` when the user wakes up or the snooze limit is reached.


**Reasoning**:
I need to complete the provided Python code in cell `dsWiDVxpbwP1` by defining a `snooze_limit`, completing the input prompt, and adding conditional logic to handle user responses for 'snooze' or 'wake up', including checking for snooze limits and invalid inputs.



```python
awake =False
snooze_count=0
snooze_limit = 3 # Define snooze_limit
while not awake:
  print("BEEP! BEEP!")
  response=input("Press s to snooze or w to wake up: ") # Complete the input prompt
  if response.lower() == 's':
    snooze_count += 1
    print("Snoozing...")
    if snooze_count >= snooze_limit:
      print("Snooze limit reached! Waking up!")
      awake = True
  elif response.lower() == 'w':
    print("Waking up!")
    awake = True
  else:
    print("Invalid response, please try again.")
```

If the user chooses to 's' (snooze), the `snooze_count` increments, and "Snoozing..." is printed. The program then checks if the `snooze_count` has reached or exceeded the `snooze_limit`. If so, it prints "Snooze limit reached! Waking up!" and sets `awake` to `True`, breaking out of the loop. This ensures that the user eventually wakes up, even if they keep snoozing.
