```python
#Checking number for even number:
num=int(input("Enter Number:"))
if num%2==0:
    print("Number is Even")
```

```python
#Checking number for even number and odd number:
num=int(input("Enter Number:"))
if num%2==0:
    print("Number is Even")
else:
    print("Number is Odd")
```

```python
#Checking positive, negative and zero number:
num=int(input("Enter Number:"))
if num>0:
    print("Number is Positive")
elif num<0:
    print("Number is Negative")
else:
    print("Number is Zero")
```

```python
#Assigning grades to the student
marks=int(input("Enter Marks:"))
if marks>=90:
    print("Grade A")
elif marks>=80:
    print("Grade B")
elif marks>=70:
    print("Grade C")
elif marks>=60:
    print("Grade D")
elif marks>=50:
    print("Grade E")
else:
    print("FAIL")
```

```python
#Assigning grades to the student using logical operators
marks=int(input("Enter Marks:"))
if marks>=90 and marks<=100:
    print("Grade A")
elif marks>=80 and marks<90:
    print("Grade B")
elif marks>=70 and marks<80:
    print("Grade C")
elif marks>=60 and marks<70:
    print("Grade D")
elif marks>=50 and marks<60:
    print("Grade E")
else:
    print("FAIL")
```

```python
#PIN Accuracy for ATM machine usage using nested if else statement:

Card_insert= True
Pin_accuracy=False
if Card_insert:
  if Pin_accuracy:
    print("You are allow to use ATM machine")
  else:
    print("You are not allow to use ATM machine because PIN incorrect")

else:
  print("You are not allow to use ATM machine because card not inserted")
```

```python
import math

# Square root
print("Square root of 25:", math.sqrt(25))
print("Square root of 2:", math.sqrt(2))
```

```python
# Ceiling and Floor
print("Ceiling of 4.3:", math.ceil(4.3))
print("Floor of 4.9:", math.floor(4.9))
print("Ceiling of -4.3:", math.ceil(-4.3))
print("Floor of -4.9:", math.floor(-4.9))
```

```python
# Factorial
print("Factorial of 5:", math.factorial(5))
print("Factorial of 0:", math.factorial(0))
```

```python
# Power
print("2 to the power of 3:", math.pow(2, 3))
print("4 to the power of 0.5 (square root of 4):", math.pow(4, 0.5))
```

```python
# Value of pi
print("Value of pi:", math.pi)
```
