Non Parameterized Functions (already done)

```python
def addition(num1,num2):
   return (num1+num2)

def subtraction(num1,num2):
   return (num1-num2)

def multiplication(num1,num2):
   return (num1*num2)

def division(num1,num2):
   if num2!=0:
    return (num1/num2)
   else:
    print("assign non zero value to num2")

added_value=addition(5,6)
print("added_value=",added_value)

substracted_value=subtraction(5,6)
print("substracted_value=",substracted_value)

multiplied_value=multiplication(5,6)
print("multiplied_value=",multiplied_value)

division_value=division(5,0)
print("division_value= ",division_value)
```

Positional Arguments

```python
def greet(name,age):
  print(f"name={name} and your age={age}")

greet("John",25)
```

```python
def greet(name,age):
  print(f"name={name} and your age={age}")

greet(25,"john")
```

```python
def greet(name,age):
  print(f"name={name} and your age={age}")

greet(age=25,name="john") #keyword argument
```

```python
def greet(name,age):
  print(f"name={name} and your age={age}")

greet(age=25,"john") #keyword argument
```
