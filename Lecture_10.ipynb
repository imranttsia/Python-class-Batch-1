#Scope of Variable

```python
name="Ayaan"
def display():
  print(name)
display()
```

```python
name="Ayaan"
def display():
  name="Ali"
  print(name)
display()
```

```python
name="Ayaan"
def display(name_name):
  print(name_name)
display(name)
```

```python
name="Ayaan" #Global variable
def display(name_name):
  new_name="Ali" #Local variable
  print(name_name)
  print(new_name)
display(name)
```

```python
name="Ayaan" #Global variable

def display(name_name):
  new_name="Ali" #Local variable
  print(name_name)
  print(new_name)

display(name)

print(name)
print(new_name)
```

```python
name="Ayaan" #Global variable

def display(name_name):
  global new_name="Ali" #Making Local variable to a global variable
  print(name_name)
  print(new_name)

display(name)

print(name)
print(new_name)
```

```python
name="Ayaan" #Global variable

def display(name_name):
  global new_name
  new="Ali" #Making Local variable to a global variable
  print(name_name)
  print(new_name)

display(name)

print(name)
print(new_name)
```

```python
name="Ayaan" #Global variable

def display(name_name):
  global new_name #Making Local variable to a global variable
  new_name="Ali"
  print(name_name)
  print(new_name)

display(name)

print(name)
print(new_name)
```
